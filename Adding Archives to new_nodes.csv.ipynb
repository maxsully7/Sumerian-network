```python
import ast
new_nodes = pd.read_csv("new_nodes.csv")
# new_nodes['p_index'][0] = [int(i) for i in new_nodes['p_index'][0] ]
new_nodes['p_index'] = [ast.literal_eval(new_nodes['p_index'][i]) for i in np.arange(len(new_nodes['p_index']))]
split = new_nodes.apply(lambda x: pd.Series(x['p_index']),axis=1).stack().reset_index(level=1, drop=True)
split
split.name = 'p_index'
split
new_df = new_nodes.drop('p_index', axis=1).join(split)
new_df['p_index'] = new_df['p_index'].apply(int)
new_df

new_df = pd.merge(new_df, archives, how='inner', left_on='p_index', right_on='PID')
new_df = new_df.drop(['PID'], axis=1)
new_df
groups = new_df.groupby(by='name').agg({'id': 'first', 
                                        'role': 'first',
                                        'profession': 'first', 
                                        'processed year': 'first', 
                                        'family': 'first', 
                                        'p_index':lambda x: list(x),
                                        'Archive': lambda x: list(x), 
                                        'date name': 'first', 
                                        'maxYear': 'first', 
                                        'minYear': 'first',
                                        'Max Gap Start Year': 'first',
                                        'Max Gap End Year': 'first'}) 
groups
# new_df = pd.merge(new_df, groups, how='inner', left_on='name', right_on='name')
# new_df = new_df.drop(['p_index_x', 'Archive_x'], axis=1)
# new_df
```

```python
bools_doa = ['domesticated animal' in groups['Archive'][i] for i in np.arange(len(groups['Archive']))]
bools_dea = ['dead animal' in groups['Archive'][i] for i in np.arange(len(groups['Archive']))]
bools_lo = ['leather object' in groups['Archive'][i] for i in np.arange(len(groups['Archive']))]
bools_po = ['precious object' in groups['Archive'][i] for i in np.arange(len(groups['Archive']))]
bools_q = ['queen' in groups['Archive'][i] for i in np.arange(len(groups['Archive']))]
bools_s = ['seal' in groups['Archive'][i] for i in np.arange(len(groups['Archive']))]
bools_u = ['unknown' in groups['Archive'][i] for i in np.arange(len(groups['Archive']))]
bools_wa = ['wild animal' in groups['Archive'][i] for i in np.arange(len(groups['Archive']))]
bools_w = ['wool' in groups['Archive'][i] for i in np.arange(len(groups['Archive']))]
groups['has domesticated animal'] = bools_doa
groups['has dead animal'] = bools_dea
groups['has leather object'] = bools_lo
groups['has precious object'] = bools_po
groups['has queen'] = bools_q
groups['has seal'] = bools_s
groups['has unknown'] = bools_u
groups['has wild animal'] = bools_wa
groups['has wool'] = bools_w
groups = groups[['id', 'role', 'profession', 'processed year', 'family', 'p_index',
       'Archive', 'has domesticated animal', 'has dead animal',
       'has leather object', 'has precious object', 'has queen', 'has seal',
       'has unknown', 'has wild animal', 'has wool', 'date name', 'maxYear', 'minYear', 'Max Gap Start Year',
       'Max Gap End Year']]
groups
```
