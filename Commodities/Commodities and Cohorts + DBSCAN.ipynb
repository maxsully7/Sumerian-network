```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.impute import SimpleImputer
from sklearn import preprocessing 
```

**Preparing the people table to get merged with the commodities**

```python
people = pd.read_csv("people_and_target.csv")
people
```

**Commodities tables (split as 1 and 2)**

```python
com1 = pd.read_csv("commodities1.csv")
com2 = pd.read_csv("commodities2.csv")
commodities = pd.concat([com1, com2])
commodities[commodities['ip'] == 125621]
```

***Commodities + People Table***

```python
people_and_comm = pd.merge(people, commodities, how = 'inner', left_on = 'p index', right_on = 'ip')
people_and_comm
# people_and_comm[people_and_comm['name'] == 'ur-mes']['p index'].unique()
```

**Run DBSCAN**

```python
people = people_and_comm.drop(['roles', 'normalized name', 'target norm_name','profession', 'family', 'Unnamed: 0', 'Unnamed: 0.1', 'p index', 'date name', 'source', 'target', "target's role"], axis = 1)
people
```

```python
imp = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
lst = []
for n in people['name'].unique():
    x = people[people['name'] == n]
    flag = False
    for c in x.columns:
        if x[c].isnull().all():
            flag = True
            break
    if flag:
        continue
    lst.append(pd.DataFrame(imp.fit_transform(x)))
```

```python
len(lst)
```

```python
people_new = pd.DataFrame()
for i in lst:
    people_new = people_new.append(i)
```

```python
people_new.columns = people.columns
people_new
```

```python
# TODO: create a LabelEncoder object and fit it to each feature in people_cat


# 1. INSTANTIATE
# encode labels with value between 0 and n_classes-1.
le = preprocessing.LabelEncoder()


# 2/3. FIT AND TRANSFORM
# use df.apply() to apply le.fit_transform to all columns
people_cat = people_new.apply(le.fit_transform)
people_cat
```

```python
# TODO: create a OneHotEncoder object, and fit it to all of people_cat

# 1. INSTANTIATE
enc = preprocessing.OneHotEncoder()

# 2. FIT
enc.fit(people_cat)

# 3. Transform
onehotlabels = enc.transform(people_cat).toarray()
onehotlabels.shape
```

```python
onehotlabels
```

```python
dates = [[people_new['processed date'].values[i]] for i in np.arange(len(people_new['processed date']))]
comm = [people_new.iloc[:, 6:len(people_new.columns)].values[i] for i in np.arange(len(people_new.iloc[:, 6:len(people_new.columns)]))]
people_cat_quant = np.hstack((onehotlabels, comm, dates))
```

```python
name_example = people_cat_quant[0:2414]
```

```python
people_new.head(2414)
# people_new[people_new['name'] == 'ur-ku₃-nun-na']
```

```python
clustering = DBSCAN(eps=2).fit(name_example)
clustering.labels_
people_2414 = people_new.head(2414)
```

```python
people_2414['label'] = clustering.labels_
people_2414
```
