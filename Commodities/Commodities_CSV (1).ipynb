```python
import pandas as pd
import re
import csv
```

```python
files = ['filtered_'+str(n)+'_with_neighbors_normalized.csv' for n in range(1,10)]
```

```python
commodities = {}
p_number = ''
comms = []
comms_list = []
adjectives = ['amar[young]N','libir[old]V/i','hulu[bad]V/i','ŋuruš[male]N']
taboo = ['ki[place]N', 'zal[pass]V/t','giŋ[unit]N', 'sila[unit]N','mana[unit]N','us[side]N', 'us[side]N',  'uri₅{ki}-ta[]GN','ib-ni-{d}iškur[]PN','bala[turn]V/t','uš[die]V/i']
for file in files:
    f = pd.read_csv(file)
    for index, row in f.iterrows():
        if row['id_word'][:7] != p_number:
            commodities[p_number] = comms
            p_number = row['id_word'][:7]
            comms = []
        w = row['lemma']
        w_pnum = row['id_word'][:7]
        if row['id_word'].count('.')==2:
            w_line = re.findall(r'\.(.*)\.',row['id_word'])[0]
        #grabs previous word to see if it's a number
            n = (f.iloc[index-1]['lemma'])
          #  print(f.iloc[index-1]['id_word'])
            if f.iloc[index-1]['id_word'].count('.')==2:
                n_line = re.findall(r'\.(.*)\.',f.iloc[index-1]['id_word'])[0]
             #   print((w_line,n_line))
                #grabs potential adjectives for checking
                if (index<(len(f)-1)):
                    adj_noun = (f.iloc[index+1]['lemma'])
                if ('(diš)[]NU' in n)and w_line==n_line:
                   # print((w_line,n_line))
                    if (w in adjectives) and adj_noun[-2:]==']N':
                        if adj_noun not in comms_list:
                            comms_list.append(adj_noun)
                        comms.append(adj_noun)
                    else:
                        if (w[-2:]==']N'):
                            if w not in comms_list:
                                comms_list.append(w)
                            if w not in taboo and w not in adjectives:
                                comms.append(w)
commodities[p_number] = [comms]
del commodities['']
```

```python
d = {'ip':list(commodities.keys()),'commodities': list(commodities.values())}
comm_df = pd.DataFrame.from_dict(data = d)
comms_list
```

```python
comm_df["#commodities"] = comm_df["commodities"].apply(lambda x:len(x))
comm_df.head(8)
```

```python
for c in comms_list:
    comm_df["% " + str(c)] = comm_df['commodities'].map(lambda a: round(a.count(c)/len(a),3) if len(a)>0 else 0)
comm_df
```
