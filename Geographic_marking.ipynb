```python
import pandas as pd
import re
import csv
```

```python
def read_pfile(file_name):
    f = open(file_name,encoding="utf8")
    ids = open('drehem_p_ids.txt', 'r')
    content = f.read().split("&P")[1:]
    g_names = {}
    content[0].split(' ')
    for text in content:
        lines = text.split('\n')
        text = text.split(' ')
        p = text[0]
        places = []
        for w in text:
            if ('{ki}' in w) or ('{ki#}' in w):
                places.append(w)
        for l in range(0, len(lines)):
            line = lines[l]
            if "GN" in line:
                line_sum = lines[l-1].split(' ')
                line_lem = line.split(' ')
                for i in range(0,len(line_lem)):
                    if line_lem[i] == "GN;":
                        if(line_sum[i] not in places):
                            places.append(line_sum[i])
        g_names[p] = [places]
    return g_names
```

```python
g_names = read_pfile('p001.txt')
geography = pd.DataFrame(data=g_names).T
geography.shape[0]
```

```python
for i in range(2,10):
    file_name = 'p00' + str(i)+ '.txt'
    file_gnames = read_pfile(file_name)
    geo_file = pd.DataFrame(data= file_gnames).T
    geography = pd.concat([geography, geo_file])
for i in range(10,16):
    file_name = 'p0' + str(i)+ '.txt'
    file_gnames = read_pfile(file_name)
    geo_file = pd.DataFrame(data= file_gnames).T
    geography = pd.concat([geography, geo_file])
```

```python
places = {}
#p_numbers = []
for g,p in zip(list(geography[0]), list(geography.index)):
    if g != []:
        for place in g:
            if place not in places.keys():
                places[place] = [p]
            else:
                places[place].append(p)
```

```python
df = pd.DataFrame(data={"names": list(places.values())})
df.to_csv("p_numbers.csv", sep=',',index=False,encoding = 'utf8')
```
