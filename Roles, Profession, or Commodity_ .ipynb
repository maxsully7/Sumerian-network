```python
#Create new professions column
filtered['prof?'] = filtered['lemma'].apply(lambda word: 'Yes' if (re.match('^[^\]]*', word)[0] + "]") in professions else 'No')
filtered['role?'] = filtered['lemma'].apply(lambda word: 'Yes' if (re.match('^[^\]]*', word)[0] + "]") in roles else 'No')
filtered['family?'] = filtered['lemma'].apply(lambda word: 'Yes' if (re.match('^[^\]]*', word)[0] + "]") in family else 'No')
```

```python
#Create "number?"" to see if row is number. this could imply that that next row is a commodity
filtered['number?'] = filtered['lemma'].str.contains('NU')
filtered['number?'] = ['Yes' if filtered['number?'][i] == True else "No" for i in range(len(filtered))]
```

```python
filtered['commodity?'] = ["No"] + ['Yes' if filtered['number?'][i-1] == "Yes" else 'No' for i in range(1, len(filtered))]
```

```python
#Check if the preceding and after row belong to the same line

pn_indices = PN.index.values.tolist()
# PN_dict = dict()
lists = [[] for i in range(len(filtered))]
for i in pn_indices:
  five_rows = filtered.loc[:i+2].tail(5)
  correct_id_line = five_rows.iloc[[2]]['id_word'].values[0][:-2]
  lemma = five_rows.iloc[[2]]['lemma'].values[0]
  #check preceding
  five_rows_some_indices = [0, 1, 3, 4]
  neighbors = []
  key = filtered.loc[[i]]['lemma'].values[0]
  for r in five_rows_some_indices:
    curr_pn = five_rows.iloc[[r]]['id_word'].values[0][:-2]
    if curr_pn == correct_id_line:
      neighbors.append(five_rows.iloc[[r]]['lemma'].values[0])
  lists[i] = neighbors
filtered['neighbors'] = lists
len(lists)
```
